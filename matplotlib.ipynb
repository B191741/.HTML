```python
x=[2,4,5,6,8,7,6,5,5]
y=[4,8,7,6,5,43,2,2,78]
```

```python
x1=[5,8,10]
y1=[10,12,13]
x2=[6,9,11]
y2=[6,12,13]
plt.bar(x1,y1,label='lineone',color='purple',linewidth='5')
plt.bar(x2,y2,label='linetwo',color='red',linewidth='3')
plt.title("figure11")
plt.legend()
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.show()
```

```python
x1=[5,8,10]
y1=[10,12,13]
x2=[6,9,11]
y2=[6,12,13]
plt.plot(x1,y1,label='lineone',color='yellow',linewidth='5')
plt.plot(x2,y2,label='linetwo',color='red',linewidth='3')
plt.title("figure11")
plt.legend()
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.show()
```
